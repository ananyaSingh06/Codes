```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
data = pd.read_csv(,encoding = 'latin1')

#data
d12 = data['day12']
d13 = data['day13']
d14 = data['day14']
d15 = data['day15']
d16 = data['day16']

d20 = data['day20']
d21 = data['day21']
d22 = data['day22']
d23 = data['day23']

d26 = data['day26']
d27 = data['day27']
d28 = data['day28']
d29 = data['day29']
d30 = data['day30']

#day
d=[d12,d13,d14,d15,d16,d20,d21,d22,d23,d26,d27,d28,d29,d30]
day=['','12','13','14','15','16','20','21','22','23','26','27','28','29','30']
pos1=np.arange(len(day))

fig=plt.figure()
ax=fig.add_axes([0,0,1.2,1.2])
ax.boxplot(d,sym='',patch_artist='True',boxprops=dict(facecolor='lightcyan'),showmeans='True')
ax.set_ylabel('Wifi Hits / Day',fontsize=12)
ax.set_xticks(pos1)
ax.set_xticklabels(day)
plt.axvline(5.5,color='black',linestyle='--',linewidth=1)
plt.axvline(9.5,color='black',linestyle='--',linewidth=1)
plt.xlabel('Day of the Month',fontsize=12)
ax.set_title('Wifi Hits',fontsize=12)
ax.yaxis.grid(True)
plt.ylim(0,100)
#plt.axhline(100,color='red',linestyle='--',linewidth=1)

# Save the figure and show

plt.show()
#plt.savefig('C:\\Users\\iaman\\OneDrive\\Desktop\\Covid Project\\Python\\NOxBP.png',bbox_inches='tight')
```

```python
fig, ay = plt.subplots()
ay=fig.add_axes([0,0,1.2,1.2])
ay.bar(pos2, mean, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10)
ay.set_ylabel('Wifi Hits / Day')
ay.set_xticks(pos2)
ay.set_xticklabels(dd)
ay.set_title('Average Daily Wifi Hits (Station A)')
ay.yaxis.grid(True)
plt.ylim(0,7000)
plt.xlabel('Day of the Month')

# Save the figure and show
plt.tight_layout()
plt.show()
```
