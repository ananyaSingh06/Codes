```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
dp = pd.read_csv(r'C:\Users\iaman\OneDrive\Desktop\SThomasGraphP.csv',encoding = 'latin1')
dw = pd.read_csv(r'C:\Users\iaman\OneDrive\Desktop\SThomasGraphW.csv',encoding = 'latin1')
```

```python
dw.pivot(index='Study Year',columns='Variable',values='A').plot(kind='bar')
plt.xlabel('Year')
plt.ylim(0,60000)
plt.ylabel('No. of Wifi Hits / Day')
plt.title('Wifi Hits April\'20 VS April\'21 (Station A)',fontsize =15)
plt.legend(title='Variable',loc='upper left')
```
